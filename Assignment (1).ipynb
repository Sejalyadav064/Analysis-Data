```python
# # 3.Matches won by minimum margins of runs
matches[matches['win_by_runs']>0].sort_values(by='win_by_runs').head(1)
```

```python
# #INSIGHTS-
# These are nail-biting matches that increase audience engagement
```

```python
 # 4.Matches won by minimum wickets
matches[matches['win_by_wickets']>0].sort_values(by='win_by_wickets').head(1) 
```

```python
# #INSIGHTS-
# * Close wicket wins create high excitement
```

```python
# 5.Matches where D/L method was applied
matches['dl_applied'].value_counts()
```

```python
# #INSIGHTS-
# Shows how many matches were affected by rain.
```

```python
# 6.Number of matches held in each city.
city_matches= matches['city'].value_counts()
plt.figure(figsize=(10,6))
city_matches.plot(kind='line')
plt.title('matches in each city')
plt.show()
```

```python
# #INSIGHTS-
# *Cities like Mumbai,Kolkata,and Banglore host more matches---> higher engagement markets.
```

```python
#7. Matches won by each team 
team_wins=matches['winner'].value_counts()
plt.figure(figsize=(10,6))
team_wins.plot(kind='line',marker='o')
plt.title('matches won by each team')
plt.show()
```

```python
# #INSIGHTS-
# *Team with consistent performance build stronger fan bases.
```

```python
# 8.Matches held every season
season_matches =matches['season'].value_counts().sort_index()
season_matches.plot(kind='barh')
plt.title("matches per season")
plt.show()
```

```python
# #INSIGHTS-
# *IPL expanded over time with more matches in later seasons.
```

```python
# 9.Top 10 Players with most man of the match
top_mom =matches['player_of_match'].value_counts().head(10)
top_mom.plot(kind='donu')
plt.title("top 10 man of the match winners")
plt.show()
```

```python
# #INSIGHTS-
# * These Players significantly impact match outcomes and brand value.
```

```python
# 10.Does winning toss mean winning match?
toss_win_match_win=matches[matches['toss_winner']==matches['winner']]
percentage =len(toss_win_match_win)/len (matches)*100
percentage
```

```python
# #INSIGHTS-
# *If percentage is around 50-55%,toss advantage is moderate, not guaranteed.
```

```python
#11.Toss decision Distribution
matches['toss_decision'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title('toss_decision distribution')
plt.show()
```

```python
# #INSIGHTS-
# *Captains prefer chasing in modern IPL seasons.
```

```python
# 12.D/L applied per season
import pandas as pd
import matplotlib.pyplot as plt
dl_season= matches[matches['dl_applied']==1]['season'].value_counts().sort_index()
dl_season.plot(kind='pie')
plt.title('D/L matches per season')
plt.show()
```
